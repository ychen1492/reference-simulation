```python
density_log = well_californie_sidetrack1_01['DT']
# micro second/foot to second/m
conversion = 1e-6/0.3048
# sandstone velocity and water velocity
v_sandstone = 5490 #m/s
v_water = 1490 #m/s
porosity = (density_log * conversion - 1./v_sandstone)/(1./v_water - 1./v_sandstone)
# write porosity to file
output_poro = {}
output_poro['PORO'] = porosity[porosity>0]
df = pd.DataFrame.from_dict(output_poro)
df.to_csv('LogData\porosity_californie_sidetrack.csv', index=False)
```

```python
temperature = pd.read_csv('./SerialResolution/temperature_resolution_dy.csv')
temperature.drop('time', inplace=True, axis=1)
# empty difference list
differences = []

# dy values
dy = []
for x in temperature.columns:
    dy.append(float(x))

for i in range(1,len(temperature.columns)):
    diff = temperature.iloc[:,i] - temperature.iloc[:,i-1]
    differences.append(sum(diff**2))

fig, ax = plt.subplots(1, 1,figsize=(10, 10))
axx = fig.axes

axx[0].plot(dy[1:], differences, linewidth=2, color='r', marker='*')
# axx[0].plot(dy[1:], np.ones(len(differences))*0.03, linestyle='dashed', color='black')
axx[0].set_xlabel('dy')
axx[0].set_xscale('log')
axx[0].set_xlim(min(dy), max(dy))
axx[0].set_ylabel('Loss')
axx[0].grid(visible=True, which='both')
```
